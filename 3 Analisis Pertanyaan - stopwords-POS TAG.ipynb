```python
from zeep import Client
from nltk.tokenize import word_tokenize
import pandas as pd
import numpy as np
import re, operator
from nltk.tag import CRFTagger
# https://yudiwbs.wordpress.com/2018/02/20/pos-tagger-bahasa-indonesia-dengan-pytho/
```

```python
# Untuk Penghilang Stopword
swClient = Client('http://fws.cs.ui.ac.id/StopwordRemover/StopwordRemover?wsdl')
# Untuk POST TAG INDO
ct = CRFTagger()
ct.set_model_file('data/all_indo_man_tag_corpus_model.crf.tagger')
```

```python
df_question_data = pd.read_csv("data/daftar_pertanyaan.csv")
question_data = list(df_question_data["Pertanyaan"])
question_data
```

```python
def pengolah_pertanyaan(inputan = ""):
    print("Pertanyaan:", inputan)
    tokenized_word=word_tokenize(inputan)
    posTag_raw = ct.tag_sents([tokenized_word])
    print("POSTagID Pertanyaan:", posTag_raw)
    
    output_stopwords = swClient.service.removeStopword(inputan)
    output_stopwords = output_stopwords.strip()
    print("Kunci pertanyaan:", output_stopwords)
    tokenized_word=word_tokenize(output_stopwords)
    posTag_stopwords = ct.tag_sents([tokenized_word])
    print("POSTagID Kunci pertanyaan:", posTag_stopwords)
    print("=====================================================\n")
```

```python
for data in question_data[3:]:
    pengolah_pertanyaan(data)
```
